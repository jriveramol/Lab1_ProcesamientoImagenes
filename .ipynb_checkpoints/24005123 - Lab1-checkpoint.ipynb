```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Problema 1

```python
# Imagen para problema 1
img = cv2.imread("ej1.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

```python
def img_reader(img, col_code):
    # lienzo en blanco
    img_out = np.zeros_like(img)
    match col_code:
        case 1:
            mask = [0, 0, 1]
        case 2:
            mask = [0, 1, 0]
        case 3:
            mask = [1, 0, 0]
        case 10:
            mask = [1, 1, 0]
        case 20:
            mask = [0, 1, 1]
        case 30:
            mask = [1, 0, 1]
        case _:
            mask = [1, 1, 1]
            
    return img*mask
```

```python
img_nueva = img_reader(img, 20)
plt.imshow(img_nueva)
plt.show()
```
